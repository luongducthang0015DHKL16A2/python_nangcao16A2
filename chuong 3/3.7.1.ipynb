# 5. Cho biết chiều cao trung bình, cân nặng trung bình của các cầu thủ. 

```python
def trung_binh(np_baseball):
      tong_weight=sum(np_baseball[:,1:])
      tong_height=sum(np_baseball[:,:1])
      so_luong=len(np_baseball)
      tb_weight=tong_weight/so_luong
      tb_height=tong_height/so_luong
      print('gia tri trung binh can nang cua vdv la:',tb_weight)
      print('gia tri trung binh can nang cua vdv la:',tb_height)
tb=trung_binh(np_baseball)
      
```
