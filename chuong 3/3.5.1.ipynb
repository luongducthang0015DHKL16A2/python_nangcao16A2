```python
import numpy as np
```

# đọc file baseball_2D.1.txt

```python
with open('baseball_2D.1.txt', 'r') as file_baseball :
      lines=file_baseball.readlines()
      
```

# lưu cột height và weight vào hai tệp txt

```python
with open('height_1.txt','w') as height_file:
      for line in lines:
            height=line.strip().split('\t')[3]
            height_file.write(height + '\n')
            
with open('weight_1.txt','w') as weight_file:
      for line in lines:
            weight=line.strip().split('\t')[4]
            weight_file.write(weight + '\n')
            
      
```

# chép dữ liệu từ tập tin vào list là height, weight,

```python
with open('height_1.txt','r') as height_1:
      height=[float(line.strip())for line in height_1.readlines()]   

with open('weight_1.txt','r') as weight_1:
      weight=[float(line.strip()) for line in weight_1.readlines()]
      
print(height)
print(weight)
```

# Tạo numpy array arr_height từ list height

```python
arr_height=np.array(height)
arr_height
```

#  Tạo numpy array arr_weight từ list weight. 

```python
arr_weight=np.array(weight)
arr_weight
```

# 3. Cho hệ số quy đổi từ inch sang m là 0.0254, tạo arr_height_m dựa trên công thức: arr_height * hệ số quy đổi. 

```python
arr_height_m=arr_height[0:]*0.0254
arr_height_m
```

# 4. Cho hệ số quy đổi từ pound sang kg là 0.453592, tạo arr_weight_kg dựa trên công thức: arr_weight * hệ số quy đổi

```python
arr_weight_kg=arr_weight[:]* 0.453592
arr_weight_kg
```

# 5. Tính BMI của arr_height_m và arr_weight_kg theo công thức BMI = Cân nặng / (Chiều cao * Chiều cao), và lưu vào arr_bmi. 

```python
arr_bmi=arr_weight_kg/(arr_height_m*arr_height_m)
arr_bmi
```
