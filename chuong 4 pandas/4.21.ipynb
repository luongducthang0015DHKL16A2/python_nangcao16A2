```python
import pandas as pd
import numpy as np
```

```python
#1. đọc dữ liệu với index_col là cột đầu tiên của dữ liệu
drink=pd.read_csv('drinks.csv')
drink
```

```python
drink.rename(columns={drink.columns[0]:'index_col'},inplace=True)
```

```python
drink
```

```python
# kiểu dữ liệu, kích thước, hiện thị tên các cột của drink
print('kiểu dữ liệu của drink là', drink.dtypes)
print('kích thước của drink là:', drink.shape)
print('tên các cột của drink:',(drink.columns))
```

```python
#cho xem 5 dòng đầu tiên và cuối cùng của drink
row5_1=pd.concat([drink.iloc[:5],drink.iloc[-1:]])
row5_1
```

```python
#5 dòng dữ liệu đầu tiên
drink.head(5)
```

```python
#dòng cuối cùng
drink.tail(1)
```

```python
#2. tính số lượng bia tiêu thụ  trung bình ở mỗi châu lục
print(drink['continent'].value_counts())
tb_beer=drink.groupby('continent')['beer_servings'].mean()
tb_beer
```

```python
#3.cho biết thông tin thống kê tổng quát(describe) số lượng rượu vang được tiêu thụ ở mỗi châu lục
describe_wine=drink.groupby('continent')['wine_servings'].describe()
describe_wine
```

```python
#4.chi biết số lượng rượu bia tiêu thụ trung bình(mean) ở mỗi châu lục
tb_beerw=drink.groupby('continent')['beer_servings','wine_servings'].mean()
tb_beerw
```

```python
#5 tính trung vị (median) của bia và rượu tiêu thụ ở mỗi châu lục
median_beerw=drink.groupby('continent')['beer_servings','wine_servings'].median()
median_beerw
```

```python
#6. số lượng rượu mạnh(spirit_servings) tiêu thụ trung bình , lớn nhất và nhỏ nhất ở mỗi châu lục
print('số lượng rượu mạnh tiêu thụ ở mỗi châu lục là')
spirit_drink=drink.groupby('continent')['spirit_servings'].mean()
spirit_drink

```

```python
print('số lượng rượu mạnh tiêu thụ thấp nhất của mỗi châu lục là')
drink.groupby('continent')['spirit_servings'].min()
```

```python
print('số lượng rượu mạnh tiêu thụ lớn nhất của mỗi châu lục là')
drink.groupby('continent')['spirit_servings'].max()
```

```python
#7.sắp xếp dữ liệu tăng dần theo số lượng bia tiêu thụ
drink_acs=drink.sort_values(by='beer_servings',ascending=True)
drink_acs
```

```python
print('5 quốc gia có số lượng tiêu thụ bia lớn nhất là',list(drink_acs['country'].tail(5)))
print('5 quốc gia có số lượng tiêu thụ bia nhỏ nhất là',list(drink_acs['country'].head(5)))
```
