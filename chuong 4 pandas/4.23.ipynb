```python
import pandas as pd
import numpy as np
```

```python
ratings=pd.read_csv('ratings.csv')
print('ratings')
ratings
```

```python
ratings.isnull().sum()
```

```python
links=pd.read_csv('links.csv')
links
```

```python
links.isnull().sum()
```

```python
# loại bỏ hàng có dữ liệu nan
links.dropna(inplace=True)
print('links')
links
```

```python
links.isnull().sum()
```

```python
movies=pd.read_csv('movies.csv')
print('movies')
movies
```

```python
movies.isnull().sum()
```

```python
tags=pd.read_csv('tags.csv')
print('tags')
tags
```

```python
tags.isnull().sum()
```

```python
all=ratings.merge(movies, on='movieId').merge(links, on='movieId').merge(tags, on='movieId')
all
```

```python
all.drop_duplicates()
```
